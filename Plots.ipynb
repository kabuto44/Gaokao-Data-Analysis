```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import csv
import operations as ops
import consts as const
```

```python
#define plotting functions
def plot_all(indicator):
    for i in range(len(const.CATEGORIES)):
        category = const.CATEGORIES[i]
        name = const.CATEGORY_NAMES[i]
        ops.singleplot(indicator,category,title=indicator+ ",\n" + name)

def save_all_lines(variable):
    for i in range(len(const.CATEGORIES)):
        category = const.CATEGORIES[i]
        name = const.CATEGORY_NAMES[i]
        short = const.SHORT_CATEGORIES[i]
        ops.saveplot(name=variable+"_"+short,variable=variable,provinces=category,title=variable+ ",\n" + name)

def scatter_all(xvar, yvar, avgx = False, avgy = False):
    for i in range(len(const.CATEGORIES)):
        category = const.CATEGORIES[i]
        name = const.CATEGORY_NAMES[i]
        ops.scatterplot(xvar, yvar,provinces=category,title=xvar + " against \n" + yvar + ",\n" + name, avgx=avgx, avgy=avgy)

def save_all_scatters(xvar,yvar, avgx = False, avgy = False):
    for i in range(len(const.CATEGORIES)):
        category = const.CATEGORIES[i]
        name = const.CATEGORY_NAMES[i]
        short = const.SHORT_CATEGORIES[i]
        ops.savescatter(name=xvar + "_" + yvar +"_"+short,xvar=xvar, yvar=yvar,provinces=category,title = yvar + " against \n" + xvar + ",\n" + name, avgx = avgx, avgy = avgy)
```

```python
ops.scatterplot("Number of Undergraduate Schools", "Undergraduate Attrition by Cohort Size", avgx = True, avgy = True)
```

```python
save_all_scatters("Number of Undergraduate Schools","Undergraduate Attrition by Cohort Size", avgx = True, avgy = True)
```
